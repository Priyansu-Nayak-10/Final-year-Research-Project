```python
import pandas as pd
```

```python
diabetes_df = pd.read_csv("../data/raw/diabetes.csv")
```

```python
print("Shape:", diabetes_df.shape)
print("\nColumns:\n", diabetes_df.columns)
print("\nMissing values:\n", diabetes_df.isnull().sum())
diabetes_df.head()
```

```python
diabetes_df.columns = diabetes_df.columns.str.lower().str.strip()
print(diabetes_df.columns)
```

```python

print(diabetes_df["diabetes"].value_counts())
print("\nData types:\n", diabetes_df.dtypes)
```

```python

categorical_cols = diabetes_df.select_dtypes(include=["object"]).columns.tolist()
numeric_cols = diabetes_df.select_dtypes(exclude=["object"]).columns.tolist()
print("Categorical:", categorical_cols)
print("Numeric:", numeric_cols)
```

```python
diabetes_df = pd.get_dummies(
    diabetes_df,
    columns=categorical_cols,
    drop_first=True
)

diabetes_df.head()
```

```python
bool_cols = diabetes_df.select_dtypes(include="bool").columns
diabetes_df[bool_cols] = diabetes_df[bool_cols].astype(int)
```
