```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python

diabetes_df = pd.read_csv("../data/raw/diabetes.csv")
heart_df = pd.read_csv("../data/raw/heartattack.csv")

diabetes_df.head()
```

```python
heart_df.head()
```

```python
print("Diabetes Info:\n")
diabetes_df.info()

print("\nHeart Attack Info:\n")
heart_df.info()
```

```python
print("Diabetes Missing Values:\n")
print(diabetes_df.isnull().sum())

print("\nHeart Attack Missing Values:\n")
print(heart_df.isnull().sum())
```

```python

diabetes_df = diabetes_df.drop(columns=[
    "HbA1c_level",
    "blood_glucose_level"
])

diabetes_df.head()
```

```python
# Drop useless / noisy columns
heart_df = heart_df.drop(columns=[
    "Patient_ID",
    "State_Name",
    "Emergency_Response_Time",
    "Annual_Income",
    "Healthcare_Access",
    "Heart_Attack_History",
    "Stress_Level",
    "Air_Pollution_Exposure",   
    "Diet_Score",  
    "Triglyceride_Level",
    "LDL_Level",
    "HDL_Level",
    "Health_Insurance"
])

# Preview after drop
heart_df.head()
```

```python
heart_df = heart_df.drop(columns=["Cholesterol_Level"])

# Preview after drop
heart_df.head()
```

```python
print("Diabetes Columns:\n", diabetes_df.columns)
print("\nHeart Columns:\n", heart_df.columns)
```

```python
diabetes_df.columns = diabetes_df.columns.str.lower()
heart_df.columns = heart_df.columns.str.lower()
```

```python
# Create encoders
le = LabelEncoder()

# Diabetes dataset
diabetes_df['gender'] = le.fit_transform(diabetes_df['gender'])
diabetes_df['smoking_history'] = le.fit_transform(diabetes_df['smoking_history'])

# Heart dataset
heart_df['gender'] = le.fit_transform(heart_df['gender'])
heart_df['smoking'] = le.fit_transform(heart_df['smoking'])
heart_df['alcohol_consumption'] = le.fit_transform(heart_df['alcohol_consumption'])
heart_df['physical_activity'] = le.fit_transform(heart_df['physical_activity'])
heart_df['family_history'] = le.fit_transform(heart_df['family_history'])
```
