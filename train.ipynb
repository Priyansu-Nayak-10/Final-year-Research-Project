```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import joblib
import os

os.makedirs("models", exist_ok=True)

# =========================
# DIABETES MODEL
# =========================

df = pd.read_csv("Dataset/diabetes.csv")

X = df[['age', 'bmi', 'bp', 'glucose']]
y = df['target']

scaler_d = StandardScaler()
X_scaled = scaler_d.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

model_d = LogisticRegression()
model_d.fit(X_train, y_train)

pred_d = model_d.predict(X_test)
print("Diabetes Accuracy:", accuracy_score(y_test, pred_d))

joblib.dump(model_d, "models/diabetes_model.pkl")
joblib.dump(scaler_d, "models/diabetes_scaler.pkl")


# =========================
# HEART MODEL
# =========================

df = pd.read_csv("Dataset/heartattack.csv")

X = df[['age', 'bp', 'cholesterol', 'smoking']]
y = df['target']

scaler_h = StandardScaler()
X_scaled = scaler_h.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

model_h = LogisticRegression()
model_h.fit(X_train, y_train)

pred_h = model_h.predict(X_test)
print("Heart Accuracy:", accuracy_score(y_test, pred_h))

joblib.dump(model_h, "models/heart_model.pkl")
joblib.dump(scaler_h, "models/heart_scaler.pkl")
```
